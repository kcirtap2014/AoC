```python
from pathlib import Path

cwd = Path.cwd()

input_file = cwd / "input.txt"
input_simple = cwd / "input_simple.txt"
input_simple2 = cwd / "input_simple2.txt"


input = input_file.read_text().splitlines()
input_simple = input_simple.read_text().splitlines()
input_simple2 = input_simple2.read_text().splitlines()
```

```python
import numpy as np
def parse_input(input):
    data = []
    
    for i, line in enumerate(input):
        subdata = np.array([x for x in line])
        data.append(subdata)
    data = np.array(data)
    rows, cols = np.where(data=="S" )
    origin = list(zip(rows, cols))[0]

    return origin, data
```

```python
def move(origin, current, data):
    if data[current] == "S":
        print("hey")
        return current[0], current[1]
    elif data[current] == "-":
        if origin[1] < current[1]:
            return current[0], current[1]+1
        elif origin[1] > current[1]:
            return current[0], current[1]-1
    elif data[current] == "|":
        if origin[0] < current[0]:
            return current[0]+1, current[1]
        elif origin[0] > current[0]:
            return current[0]-1, current[1]
    elif data[current] == "L":
        if origin[0] < current[0]:
            return current[0], current[1]+1
        elif origin[1] > current[1]:
            return current[0]-1, current[1]
    elif data[current] == "F":
        if origin[0] > current[0]:
            return current[0], current[1]+1
        elif origin[1] > current[1]:
            return current[0]+1, current[1]
    elif data[current] == "J":
        if origin[0] < current[0]:
            return current[0], current[1]-1
        elif origin[1] < current[1]:
            return current[0]-1, current[1]
    elif data[current] == "7":
        if origin[0] > current[0]:
            return current[0], current[1]-1
        elif origin[1] < current[1]:
            return current[0]+1, current[1]
    
```

```python
def initialise(data, origin):
    row,col = data.shape

    if origin[0] + 1 < row:
        origin_data = data[origin[0]+1, origin[1]]
        if origin_data in ["J", "|", "L"] :
            current = (origin[0]+1, origin[1])
            return current
        
    if origin[0] - 1 >= 0:
        origin_data = data[origin[0]-1, origin[1]]
        if origin_data in ["7", "|", "F"]:
            current = (origin[0]-1, origin[1])
            return current
    
    if origin[1] + 1 < col:
        origin_data = data[origin[0], origin[1]+1]
        if origin_data in ["J", "-", "7"]:
            current = (origin[0], origin[1]+1)
            return current
    
    if origin[1] - 1 >= 0:
        origin_data = data[origin[0], origin[1]-1]
        if origin_data in ["L", "-", "F"]:
            current = (origin[0], origin[1]-1)
            return current
```

```python
def main(data, origin):
    i=0
    move_array = []
    current = initialise(data, origin)
    move_array.append(current)
    
    while np.logical_or(data[current]!="S", i==0):
        inter = current
        current = move(origin, current, data)
        move_array.append(current)
        origin = inter
        i+=1
        
    return move_array
    
```
