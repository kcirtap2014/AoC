```python
from pathlib import Path
cwd = Path.cwd()
input_text = (cwd / 'input_d2.txt').read_text()
input_text_simple = (cwd / 'input_d2_simple.txt').read_text()
```

```python
from collections import defaultdict
constraint = {"red":12, "green":13, "blue":14}
def parse_input(input_text, constraint ):
    id_num = []
    sum_id = 0
    for k, line in enumerate(input_text.splitlines()):
        
        semicolon = line.split(": ")[-1].split("; ")
        semi_valid = []
        
        print("new")
        for semi in semicolon:
            spl_dict = defaultdict(int)
            spl = semi.split(", ")
            for s in spl:
                ss = s.split(" ")
                spl_dict[ss[1].strip()] += int(ss[0].strip())
            in_valid = []
            for key, value in spl_dict.items():
                if value > constraint[key]:
                    in_valid.append(False)
                else:
                    in_valid.append(True)
            semi_valid.append(all(in_valid))
            print(semi_valid)
        if all(semi_valid)==True:
            id_num.append(k+1)
            sum_id += k+1
        
    return id_num, sum_id
```

```python
id_nums, sum_ids = parse_input(input_text, constraint )
```
