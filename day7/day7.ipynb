```python
def get_dominant_key(cards):
    return sorted(cards, key=lambda x: cards[x], reverse=True)[0]
```

```python
import numpy as np
def check_type(cards):
    values = np.array(list(cards.values()))
    
    if len(np.where(values==5)[0])>0:
        key = list(cards.keys())[list(cards.values()).index(5)]
        ans=5
    elif len(np.where(values==4)[0])>0:
        ans=4
        key = list(cards.keys())[list(cards.values()).index(4)]
    elif len(np.where(np.logical_and(3 in values==3, 2 in values))[0])>0:
        ans=3
        key = list(cards.keys())[list(cards.values()).index(3)]
    elif len(np.where(values==3)[0])>0:
        ans=2
        key =list(cards.keys())[list(cards.values()).index(3)]
    elif len(np.where(values==2)[0])>1:
        ans = 1
        key = list(cards.keys())[list(cards.values()).index(2)]
    elif len(np.where(values==2)[0])==1:
        ans=0
        key = list(cards.keys())[list(cards.values()).index(2)]
    elif len(np.where(values==1)[0])==5:
        ans = -1
        key = -1
    else:
        print("problem")
        
    return {'ans':ans, 'key':key}

            
        
```

```python
from collections import defaultdict

def count_card(card,  joker=False):
    cards = defaultdict(int)
    for c in card:
        cards[c]+=1

    if joker:
        dom_key = get_dominant_key(cards)
        if dom_key != "J":
            cards[dom_key] += cards["J"]
            cards["J"] = 0
        else:
            if cards["J"] !=5:
                cards[list(cards.keys())[0]] += cards["J"]
                cards["J"] = 0
    return cards

def parse_input(input, order_string, joker):
    data = defaultdict(dict)

    for k, line in enumerate(input):
        card = line.split()[0]
        score = line.split()[1]
        sorted_card = "".join(sorted(card, key=lambda x: order_string.index(x)))
        data[k]["card"] = check_type(count_card(sorted_card, joker))
        data[k]["hand"] = card
        data[k]["score"] = int(score)

    return data
```

```python
dict_out = parse_input(input, order_string=order_string_J, joker=True)
```
