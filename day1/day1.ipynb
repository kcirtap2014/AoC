```python
import re
def parse_input(input_text):
    double_digits = []
    for line in input_text.splitlines():
        all_digits = "".join(re.findall(r"\d+", line))
        double_digit = int(all_digits[0] + all_digits[-1])
        double_digits.append(double_digit)
    return double_digits
```

```python
import re

def get_all_digits(sentence):
    digits = re.findall(r'\d+|one|two|three|four|five|six|seven|eight|nine', sentence)
    print(digits)
    digits = [str(d) for d in digits if d.isdigit()]
    return ''.join(digits)
```

```python
# Part 2
dict = {"one":1, "two":2, "three":3,"four":4, "five":5, "six":6, "seven":7, "eight":8, "nine":9}
import re
def parse_input2(input_text):
    double_digits = []
    sum_all = 0
    for line in input_text.splitlines():
        all_digits = re.findall(r"\d+|one|two|three|four|five|six|seven|eight|nine", line)
        new_digits = []
        for ad in all_digits:
            if ad in dict.keys():
                new_digits.append(dict[ad])
            else:
                new_digits.append(ad)
            all_digits = "".join(map(str, new_digits))
        double_digit = int(all_digits[0] + all_digits[-1])
        double_digits.append(double_digit)
        sum_all+=double_digit
    return sum_all
```

```python

sum = 0
lookups = ["[0-9]", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
with open("input.txt", 'r') as input:
    while line := input.readline():
        numbers = re.findall(f"(?=({'|'.join(lookups)}))", line)
        first   = numbers[0]                if numbers[0].isdigit()                else lookups.index(numbers[0])
        second  = numbers[len(numbers) - 1] if numbers[len(numbers) - 1].isdigit() else lookups.index(numbers[len(numbers) - 1])
        
        sum += int(f"{first}{second}")
        print(int(f"{first}{second}"))
print("Sum: {0}".format(sum))
```
