```python
from pathlib import Path

cwd = Path.cwd()

input_file = cwd / "input.txt"
input_simple = cwd / "input_simple.txt"

input = input_file.read_text().splitlines()
input_simple = input_simple.read_text().splitlines()
```

```python
import numpy as np
def find_diff_extrapolate(data, first=False):
    diff = np.array(data)

    if first:
        last_num = [diff[0]]
        extrapolate_list = []
    else:
        last_num = [diff[-1]]
        extrapolate_list =[diff[-1]]
        
    while np.sum(diff) != 0:
        diff = np.diff(diff)
        if first:
            last_num.append(diff[0])
           
        else:
            last_num.append(diff[-1])

    if first:
        extrapolate_list.append(last_num[-1])
        for i in range(len(last_num)-1, 0, -1):
            #print(extrapolate_list[len(last_num) - i-1] , last_num[i-1])
            extrapolate_list.append( last_num[i-1] - extrapolate_list[len(last_num) - i-1])
    else:
        for i in range(len(last_num)-1, 1, -1):
            extrapolate_list.append(extrapolate_list[len(last_num) - i-1] + last_num[i-1])

    return  extrapolate_list
```

```python
def parse_input(input, first=False):
    sum_a = 0
    extra = []
    for line in input:
        data = np.array([int(x) for x in line.split(" ")])
        extrapolate = find_diff_extrapolate(data, first=first)
        extra.append(extrapolate[-1])
        sum_a+= extrapolate[-1]
    return sum_a,extra
```

```python
sum_a, extra = parse_input(input, first=True)
```
